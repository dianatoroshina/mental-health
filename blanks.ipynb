```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv("MENTAL HEALTH DATASET.csv")
data
```

```python
print(data.info())
```

```python
data["Gender"].unique()
```

```python
def clean_gender(gender):
    gender = gender.lower().strip()
    
    if any(keyword in gender for keyword in ['female', 'woman', 'f', 'cis female']):
        return 'female'
    
    elif any(gen in gender for gen in ['male', 'man', 'm', 'cis male', 'guy']):
        return 'male'
    
    else:
        return 'other'
```

```python
data['Gender'] = data['Gender'].apply(clean_gender)
print(data['Gender'])
```

```python
print('Значения колонки state: \n', data['state'].unique(),'\n',
      'Количество отсутствующих значений: \n', data['state'].isna().sum())
```

Так как отсутствует больше 40% данных, столбец state можно удалить

```python
data.drop(labels='state', axis=1, inplace=True)
```

```python
columns_yesno = [
                 'self_employed', 'family_history', 'treatment',
                 'remote_work', 'tech_company', 'benefits',
                 'care_options', 'wellness_program', 'seek_help', 
                 'anonymity', 'leave',
                 'mental_health_consequence', 'phys_health_consequence',
                 'coworkers', 'supervisor', 'mental_health_interview', 
                 'phys_health_interview', 'mental_vs_physical', 'obs_consequence'
                ]

for column in columns_yesno:
    print(f"Количество уникальных и пропущенных значений в {column}:", data[column].unique(), data[column].isna().sum())
```

Отсутствующие значения присутствуют только в колонке self_employed. Заполним чаще встречающимися значениями

```python
data['self_employed'].fillna(data['self_employed'].mode()[0], inplace=True)
data
```

```python
for column in ['work_interfere', 'no_employees']:
    print(f"Количество уникальных и пропущенных значений в {column}:", data[column].unique(), data[column].isna().sum())
    
print("Количество пропущенных значений в comments:", data['comments'].isna().sum())
comments = pd.Series(data['comments']) # Оставим на дальнейшее отдельное рассмотрение
data.drop(labels='comments', axis=1, inplace=True)
```

```python
data['work_interfere'].fillna(data['work_interfere'].mode()[0], inplace=True)
data
```

```python
Q1 = data['Age'].quantile(0.25)
Q3 = data['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['Age'] < lower_bound) | (data['Age'] > upper_bound)]
print(f"Количество выбросов: {len(outliers)}")
```

```python
data_cleaned = data[~((data['Age'] < lower_bound) | (data['Age'] > upper_bound))]

print(f"Размер данных до удаления выбросов: {data.shape}")
print(f"Размер данных после удаления выбросов: {data_cleaned.shape}")
```

```python
data_cleaned.to_csv('data_processed.csv', index=False)
```
